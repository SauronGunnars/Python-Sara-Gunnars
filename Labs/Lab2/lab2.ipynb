```python
#Read the data and saved in varible "data"
path1 = "../Lab2/datapoints.txt"
lines = []

width = []
height = []
pokemon = []

with open(path1, "r") as f_data:

    lines = [line.rstrip() for line in f_data]  # saves each line as a list
    del lines[0] #deletes the first line

    # converts to float --> splits element into seperatate lists
    width = [float(element.split(",")[0]) for element in lines]
    height = [float(element.split(",")[1]) for element in lines]
    pokemon = [float(element.split(",")[2]) for element in lines]

import matplotlib.pyplot as plt #import matplot in order to plot diagram

plt.scatter(width, height, c= pokemon)
plt.xlabel("Pokemon width");
plt.ylabel("Pokemon height");
```
